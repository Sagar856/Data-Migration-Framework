```sql
update
  control.processqueue
set
  ProcessStatus = 'ReadyforProcessing', QueuedTime = current_timestamp()
where
  ProcessStatus = 'AwaitingProcessing'
  and StageId = 1;

select
  TableName,
  ProcessInstanceId,
  ProcessQueueId,
  StageId
from
  control.processqueue
where
  ProcessStatus = 'ReadyforProcessing'
  and StageId = 1;
```
